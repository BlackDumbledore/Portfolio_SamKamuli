```python
unique_host = df['hostname'].nunique()
print(unique_host)
#It is clear from our dataset, there are 26 unique host stars.
```

```python
planets_per_star = df.groupby('hostname')['pl_name'].nunique()
print(planets_per_star.head(10))
```

```python
moons_per_planet = df.groupby('pl_name')['sy_mnum'].sum().reset_index()
print(moons_per_planet.head(10))
#As at now, there have been no known moons orbiting any exoplanets.
```

```python
circumbinary_planets = df[df['cb_flag'] == 1][['pl_name', 'hostname']]
circumbinary_planets = circumbinary_planets.reset_index(drop=True)
print("Circumbinary Planets and Their Host Stars:")
print(circumbinary_planets)
#So far, 2MASS J01033563-5515561 AB b has been identified as the only circumbinary planet
```

```python
df_unique = df.drop_duplicates(subset='pl_name', keep='first')
planets_by_year = df_unique.groupby('disc_year')['pl_name'].count().reset_index()
planets_by_year.columns = ['Discovery Year', 'Number of Planets']
planets_by_year = planets_by_year.sort_values(by='Discovery Year')
print(planets_by_year.head(5))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(planets_by_year['Discovery Year'], planets_by_year['Number of Planets'], marker='o')
plt.title('Number of Exoplanets Discovered Each Year')
plt.xlabel('Discovery Year')
plt.ylabel('Number of Planets')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(planets_by_year['Discovery Year'], planets_by_year['Number of Planets'],color='deepskyblue', alpha=0.7, edgecolors='black')
import numpy as np
z = np.polyfit(planets_by_year['Discovery Year'], planets_by_year['Number of Planets'], 1)
p = np.poly1d(z)
plt.plot(planets_by_year['Discovery Year'], p(planets_by_year['Discovery Year']), "r--", label='Trend Line')
plt.title('Exoplanet Discoveries by Year (NASA Exoplanet Archive)')
plt.xlabel('Discovery Year')
plt.ylabel('Number of Planets Discovered')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
#There has been a slight upward trend over 25 years in number of Exoplanets discovered.
```

```python
df_unique = df.drop_duplicates(subset=['pl_name','disc_facility'], keep='first')
planet_by_facility = df_unique.groupby('disc_facility')['pl_name'].count().reset_index()
planet_by_facility = planet_by_facility.sort_values(by='pl_name', ascending=False)
print(planet_by_facility.head)
```

```python
df_unique2=df.drop_duplicates(subset=['pl_name','disc_telescope'], keep='first')
planet_by_telescope=df_unique2.groupby('disc_telescope')['pl_name'].count().reset_index()
planet_by_telescope=planet_by_telescope.sort_values(by='pl_name', ascending= False)
print(planet_by_telescope)
```

```python
avg_mass_per_planet = df.groupby('pl_name', as_index=False)['pl_bmasse'].mean()
avg_mass_per_planet = avg_mass_per_planet.merge(
    df[['pl_name', 'hostname']].drop_duplicates(),
    on='pl_name',
    how='left'
)
avg_mass_per_planet = avg_mass_per_planet[
    avg_mass_per_planet['pl_bmasse'] > 0
]
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.scatter(
    avg_mass_per_planet['hostname'],
    avg_mass_per_planet['pl_bmasse'],
    alpha=0.7
)
plt.xticks(rotation=90)
plt.xlabel('Host Star')
plt.ylabel('Average Planet Mass (Earth Masses)')
plt.title('Average Planet Mass per Host Star (Excluding Zero Mass Values)')
plt.tight_layout()
plt.show()
```
